```python
spark = SparkSession.builder \
    .appName("YourAppName") \
    .master("local[*]").getOrCreate()
```

```python
num_clusters = 5
```

```python
cluster_data = {}
```

```python
schema = StructType([
    StructField("status_id", IntegerType(), nullable=True),
    StructField("status_type", StringType(), nullable=True),
    StructField("status_published", StringType(), nullable=True),
    StructField("num_reactions", IntegerType(), nullable=True),
    StructField("num_comments", IntegerType(), nullable=True),
    StructField("num_shares", IntegerType(), nullable=True),
    StructField("num_likes", IntegerType(), nullable=True),
    StructField("num_loves", IntegerType(), nullable=True),
    StructField("num_wows", IntegerType(), nullable=True),
    StructField("num_hahas", IntegerType(), nullable=True),
    StructField("num_sads", IntegerType(), nullable=True),
    StructField("num_angrys", IntegerType(), nullable=True)])
```

```python
for cluster_id in range(num_clusters):
    file_path = os.path.join(MAIN_PATH, f'cluster{cluster_id}.csv')
    cluster_data[cluster_id+1] = spark.read.option("header", True).csv(file_path,schema = schema)
```

```python
for cluster_id, cluster_df in cluster_data.items():
    print(f"\nCluster {cluster_id} Statistics:")
    cluster_df = cluster_df.withColumn('status_published', F.to_timestamp('status_published'))
    cluster_df = (cluster_df
      .withColumn('year_month', F.date_format('status_published', 'yyyy-MM'))
      .withColumn('hour', F.hour('status_published'))
      .withColumn('hour_bin', ((F.col('hour') / 4).cast('int') * 4).cast('int'))
     )
    reaction_columns = ['num_comments', 'num_shares', 'num_likes', 'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys']

    # Group by 'year_month', 'hour_bin', and 'status_type' and calculate the mean of 'num_reactions'
    grouped_data = cluster_df.groupBy('hour_bin', 'status_type').agg(*[F.mean(col).alias(f'mean_{col}') for col in reaction_columns])

    # Pivot the table for better visualization
    pivot_data = (grouped_data
                  .groupBy('hour_bin')
                  .pivot('status_type')
                  .agg(*[F.first(f'mean_{col}').alias(f'{col}') for col in reaction_columns])
                  .fillna(0)
                )


    # Convert PySpark DataFrame to Pandas for plotting
    pivot_data_pd = pivot_data.toPandas()
    pivot_data_pd = pivot_data_pd.sort_values(by=['hour_bin'])

    # Plotting
    plt.figure(figsize=(15, 8))
    sns.heatmap(pivot_data_pd.set_index(['hour_bin']), cmap='YlGnBu', linewidths=.5, annot=True, fmt=".1f", cbar_kws={'label': 'Num Reactions'})
    plt.title('Mean Num Reactions Based on Publish Time (4-hour bins) and Status Type')
    plt.xlabel('Status Type')
    plt.ylabel('4-Hour Bins')
    plt.show()
```

```python
for cluster_id, cluster_df in cluster_data.items():
    print(f"\nCluster {cluster_id} Statistics:")
    cluster_df = cluster_df.withColumn('status_published', F.to_timestamp('status_published'))
    cluster_df = (cluster_df
      .withColumn('year_month', F.date_format('status_published', 'yyyy-MM'))
      .withColumn('hour', F.hour('status_published'))
      .withColumn('hour_bin', ((F.col('hour') / 4).cast('int') * 4).cast('int'))
     )
    reaction_columns = ['status_type']

    # Group by 'hour_bin', 'status_type' and calculate the count of data points
    grouped_data = cluster_df.groupBy('hour_bin', 'status_type').agg(*[F.count(col).alias(f'count_{col}') for col in reaction_columns])

    # Pivot the table for better visualization
    pivot_data = (grouped_data
                  .groupBy('hour_bin')
                  .pivot('status_type')
                  .agg(*[F.first(f'count_{col}').alias(f'{col}') for col in reaction_columns])
                  .fillna(0)
                )

    # Convert PySpark DataFrame to Pandas for plotting
    pivot_data_pd = pivot_data.toPandas()
    pivot_data_pd = pivot_data_pd.sort_values(by=['hour_bin'])

    # Plotting
    plt.figure(figsize=(15, 8))
    sns.heatmap(pivot_data_pd.set_index(['hour_bin']), cmap='YlGnBu', linewidths=.5, annot=True, fmt="g", cbar_kws={'label': 'Num Data Points'})
    plt.title('Count of Data Points Based on Publish Time (4-hour bins) and Status Type')
    plt.xlabel('Status Type')
    plt.ylabel('4-Hour Bins')
    plt.show()
```
